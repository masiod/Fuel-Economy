# Merging Datasets
Use pandas Merges to create a combined dataset from `clean_08.csv` and `clean_18.csv`. You should've created these data files in the previous section: *Fixing Data Types*.

```python
# load datasets
import pandas as pd
df_08 = pd.read_csv('./clean_08.csv')
df_18 = pd.read_csv('./clean_18.csv')
```

```python
#display first five rows on dataset
df_08.head()
```

```python
#loocking to know columns data type and null values
df_08.info()
```

```python
#loocking to know columns data type and null values
df_18.info()
```

### Create combined dataset

```python
# rename 2008 columns
df_08.rename(lambda x: x[:10] + "_2008",axis='columns',inplace=True)
```

```python
# view to check names
df_08.head()
```

```python
# merge datasets
df_combined = pd.merge(df_08,df_18,left_on='model_2008',right_on='model')
```

```python
# view to check merge
df_combined.head()
```

Save the combined dataset

```python
df_combined.to_csv('combined_dataset.csv', index=False)
```

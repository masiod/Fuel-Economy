# Merging Datasets
Use pandas Merges to create a combined dataset from `clean_08.csv` and `clean_18.csv`. You should've created these data files in the previous section: *Fixing Data Types*.

```python
# load datasets
import pandas as pd
df_08 = pd.read_csv('./clean_08.csv')
df_18 = pd.read_csv('./clean_18.csv')
```

### Create combined dataset

```python
# rename 2008 columns
df_08.rename(lambda x: x[:10] + "_2008",axis='columns',inplace=True)
```

```python
# view to check names
df_08.head()
```

```python
# merge datasets
df_combined = pd.merge(df_08,df_18,left_on='model_2008',right_on='model')
```

```python
df_combined.head(10)
```

```python
# we have to use group by func model and find the mean cmb_mpg_2008 and mean cmb_mpg for each.
model_mpg=df_combined.groupby('model').mean(numeric_only=True)[['cmb_mpg_2008','cmb_mpg']]
```

```python
model_mpg.head()
```

```python
#we will creat new column, `mpg_change`, with the change in mpg 
model_mpg['mpg_change'] = model_mpg['cmb_mpg'] - model_mpg['cmb_mpg_2008']
model_mpg.head()
```

```python
#Find the vehicle that improved the most
max_change = model_mpg['mpg_change'].max()
model_mpg[model_mpg['mpg_change'] == max_change]
```

```python
#use idxmax to find index of the row containing a column's maximun value
idx = model_mpg.mpg_change.idxmax()

idx
```
